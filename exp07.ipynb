```python
# example 7

import numpy as np
import matplotlib.pyplot as plt

# peaks funtion
def peaks(x,y):
    return 3*(1-x)**2*np.exp(-(x**2)-(y+1)**2) \
           -10*(x/5-x**3-y**5)*np.exp(-x**2-y**2) \
           -1/3*np.exp(-(x+1)**2-y**2)

# data
x = np.arange(-5.0,5.0,0.2)
y = np.arange(-5.0,5.0,0.2)
X, Y = np.meshgrid(x,y)
Z = peaks(X,Y)
V, U = np.gradient(Z,0.25)

# matplotlib
fig1 = plt.figure(figsize=(5,3),dpi=100,facecolor='w',edgecolor='k')
plt.quiver(X,Y,U,V,color='b')
plt.xlim(-5,5); plt.ylim(-5,5)
plt.title('Quiver')
plt.savefig('exp07a.png',dpi=100,bbox_inches='tight')
plt.show()
```

```python
fig2 = plt.figure(figsize=(5,3),dpi=100,facecolor='w',edgecolor='k')
sp = 15; ep = 25
plt.quiver(X[sp:ep,sp:ep],Y[sp:ep,sp:ep],U[sp:ep,sp:ep],V[sp:ep,sp:ep],color='b')
plt.title('Zoom in')
plt.savefig('exp07b.png',dpi=100,bbox_inches='tight')
plt.show()
```
