```python
# example 1

import numpy as np
import matplotlib.pyplot as plt

# data
x = np.linspace(0,10,num=21)
y1 = np.exp(x/2); e1 = 10*np.random.rand(y1.shape[0],)
y2 = np.exp(x/3); e2 =  8*np.random.rand(y2.shape[0],)
y3 = np.exp(x/4); e1 =  6*np.random.rand(y3.shape[0],)
y4 = np.exp(x/5); e2 =  4*np.random.rand(y4.shape[0],)

# plot
fig = plt.figure(figsize=(5,3),dpi=100,facecolor='w',edgecolor='k')
plt.plot(x,y1,linestyle='-',linewidth=1)
plt.plot(x,y2,linestyle='-.',linewidth=1)
plt.plot(x,y3,linestyle='--',linewidth=1)
plt.plot(x,y4,linestyle=':',linewidth=1)
plt.xlim(0,10)
plt.ylim(0,150)
plt.xlabel("x",fontsize=12)
plt.ylabel("y",fontsize=12)
plt.savefig('exp01a.png',dpi=100,bbox_inches='tight')
```

```python
fig = plt.figure(figsize=(5,3),dpi=100,facecolor='w',edgecolor='k')
plt.plot(x,y1,'ko',markeredgecolor='k',markersize=5,linewidth=1)
plt.plot(x,y2,'bs',markeredgecolor='k',markersize=5,linewidth=1)
plt.plot(x,y3,'rd',markeredgecolor='k',markersize=5,linewidth=1)
plt.plot(x,y4,'g^',markeredgecolor='k',markersize=5,linewidth=1)
plt.xlim(0,10)
plt.ylim(0,150)
plt.xlabel("x",fontsize=12)
plt.ylabel("y",fontsize=12)
plt.savefig('exp01b.png',dpi=100,bbox_inches='tight')
```

```python
fig = plt.figure(figsize=(10,6),dpi=100,facecolor='w',edgecolor='k')
ax1 = plt.subplot(2,2,1)
plt.plot(x,y1,linestyle='-',linewidth=1)
plt.plot(x,y2,linestyle='-.',linewidth=1)
plt.plot(x,y3,linestyle='--',linewidth=1)
plt.plot(x,y4,linestyle=':',linewidth=1)
plt.xlim(0,10)
plt.ylim(0,150)
plt.ylabel("y",fontsize=12)

ax2 = plt.subplot(2,2,2)
plt.plot(x,y1,'k-',linewidth=1)
plt.plot(x,y2,'b-.',linewidth=2)
plt.plot(x,y3,'r--',linewidth=3)
plt.plot(x,y4,'g:',linewidth=4)
plt.xlim(0,10)
plt.ylim(0,150)

ax3 = plt.subplot(2,2,3)
plt.plot(x,y1,'ko-',markersize=2,linewidth=1)
plt.plot(x,y2,'bs-',markersize=3,linewidth=1)
plt.plot(x,y3,'rd-',markersize=4,linewidth=1)
plt.plot(x,y4,'g^-',markersize=5,linewidth=1)
plt.xlim(0,10)
plt.ylim(0,150)
plt.xlabel("x",fontsize=12)
plt.ylabel("y",fontsize=12)

ax4 = plt.subplot(2,2,4)
plt.plot(x,y1,'ko',markeredgecolor='k',markersize=5,linewidth=1)
plt.plot(x,y2,'bs',markeredgecolor='k',markersize=5,linewidth=1)
plt.plot(x,y3,'rd',markeredgecolor='k',markersize=5,linewidth=1)
plt.plot(x,y4,'g^',markeredgecolor='k',markersize=5,linewidth=1)
plt.xlim(0,10)
plt.ylim(0,150)
plt.xlabel("x",fontsize=12)
plt.savefig('exp01c.png',dpi=100,bbox_inches='tight')
```
