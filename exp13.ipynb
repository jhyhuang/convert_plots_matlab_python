```python
# example 13

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.animation as animation
from IPython.display import HTML

def update(num,x,y,z,r,plot):
    plot[0].remove()
    x_ = r[num]*x
    y_ = r[num]*y
    z_ = r[num]*z
    plot[0] = ax.plot_surface(x_,y_,z_,cmap="twilight")

# data
n = 30
phi = np.linspace(0,2*np.pi,n)
theta = np.linspace(0,np.pi,n)
r = np.hstack((np.linspace(0,10,int(n/2)),np.linspace(10,0,int(n/2))))
x = np.outer(np.cos(phi),np.sin(theta)) 
y = np.outer(np.sin(phi),np.sin(theta))
z = np.outer(np.ones(np.size(phi)),np.cos(theta))
ticks = np.arange(-10,10,5) 

# example figure
fig = plt.figure(figsize=(6,6),dpi=100,facecolor='w',edgecolor='k')
ax = fig.gca(projection='3d')
ax.set_xlim(-10.1,10.1); ax.set_ylim(-10.1,10.1); ax.set_zlim(-10.1,10.1)
ax.set_xticks(ticks); ax.set_yticks(ticks); ax.set_zticks(ticks);
ax.set_xlabel("x"); ax.set_ylabel("y"); ax.set_zlabel("z")
ax.plot_surface(r[10]*x,r[10]*y,r[10]*z,cmap="twilight");
fig.savefig('exp13a.png',dpi=100,bbox_inches='tight')
```

```python
# example figures
fig = plt.figure(figsize=(10,6),dpi=100,facecolor='w',edgecolor='k')
for i in range(0,6):
    ax = fig.add_subplot(2,3,i+1, projection='3d')
    ax.set_xlim(-10.1,10.1); ax.set_ylim(-10.1,10.1); ax.set_zlim(-10.1,10.1)
    ax.set_xticks([]); ax.set_yticks([]); ax.set_zticks([])
    if i==0:
        ax.plot_surface(r[1]*x,r[1]*y,r[1]*z,cmap="twilight");
    else:
        ax.plot_surface(r[3*i]*x,r[3*i]*y,r[3*i]*z,cmap="twilight");
fig.savefig('exp13b.png',dpi=100,bbox_inches='tight')
```
