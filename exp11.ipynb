```python
# example 11

import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objs as go

# peaks funtion
def peaks(x,y):
    return 3*(1-x)**2*np.exp(-(x**2)-(y+1)**2) \
           -10*(x/5-x**3-y**5)*np.exp(-x**2-y**2) \
           -1/3*np.exp(-(x+1)**2-y**2)

# data
x = np.arange(-5.0,5.0,0.2)
y = np.arange(-5.0,5.0,0.2)
X, Y = np.meshgrid(x, y)
Z = peaks(X,Y)

# matplotlib: contour
fig1 = plt.figure(figsize=(6,5),dpi=100)
ax1 = fig1.gca(projection='3d')
cont = ax1.contour(X,Y,Z,20)
ax1.set_xlim(-5,5); ax1.set_ylim(-5,5); ax1.set_zlim(-8,8)
ax1.set_title('Contour lines',fontsize=12)
fig1.colorbar(cont, shrink=0.5, aspect=5)
plt.savefig('exp11a.png',dpi=100)
```

```python
# matplotlib: surface 
fig2 = plt.figure(figsize=(6,5),dpi=100)
ax2 = fig2.gca(projection='3d')
surf = ax2.plot_surface(X,Y,Z,cmap='seismic')
ax2.set_xlim(-5,5); ax2.set_ylim(-5,5); ax2.set_zlim(-8,8)
ax2.set_title('Surface',fontsize=12)
fig2.colorbar(surf, shrink=0.5, aspect=5)
plt.savefig('exp11b.png',dpi=100)
```

```python
# plotly
surf = go.Surface(x=X,y=Y,z=Z,colorscale='balance')
data = [surf]
fig3 = go.Figure(data=data)
camera = dict(eye=dict(x=2,y=2,z=1))
fig3.update_layout(scene_camera=camera,width=600,height=500)
fig3.show()
```
