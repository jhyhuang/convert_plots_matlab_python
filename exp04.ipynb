```python
# example 4

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# data
data = sns.load_dataset("mpg")

# plots
cmap = ["b","r","y"]
fig = plt.figure(figsize=(5,3),dpi=100,facecolor='w',edgecolor='k')
k = 0
for i in [8,6,4]:
    data_x = data[(data['cylinders']==i)]['horsepower']
    data_y = data[(data['cylinders']==i)]['mpg']
    plt.scatter(data_x,data_y,color=cmap[k],alpha=0.5)
    k = k+1
plt.xlim(50,230); plt.ylim(0,50)
plt.xlabel('Horsepower',fontsize=10)
plt.ylabel('MPG',fontsize=10)
plt.legend(['8 cylinders','6 cylinders','4 cylinders'])
plt.savefig('exp04a.png',dpi=100,bbox_inches='tight')
```

```python
fig = plt.figure(figsize=(10,3),dpi=100,facecolor='w',edgecolor='k')
ax1 = plt.subplot(1,2,1)
k = 0
for i in [8,6,4]:
    data_x = data[(data['cylinders']==i)]['horsepower']
    plt.hist(data_x,bins=50,color=cmap[k],alpha=0.5)
    k = k+1
plt.ylabel('Horsepower',fontsize=10)
plt.legend(['8 cylinders','6 cylinders','4 cylinders'])
plt.grid(True)

ax2 = plt.subplot(1,2,2)
k = 0
for i in [8,6,4]:
    data_y = data[(data['cylinders']==i)]['mpg']
    plt.hist(data_y,bins=50,color=cmap[k],alpha=0.5)
    k = k+1
plt.ylabel('MPG',fontsize=10)
plt.legend(['8 cylinders','6 cylinders','4 cylinders'])
plt.grid(True)
plt.savefig('exp04b.png',dpi=100,bbox_inches='tight')
```

```python
fig = plt.figure(figsize=(5,5),dpi=100,facecolor='w',edgecolor='k')
sub_data = data[(data['cylinders']==8) | (data['cylinders']==6) | (data['cylinders']==4)]
sns.jointplot(x="horsepower",y="mpg",kind="scatter",hue="cylinders",data=sub_data)
plt.savefig('exp04c.png',dpi=100,bbox_inches='tight')
```
