```python
# example 10

import numpy as np
import matplotlib.pyplot as plt

# data - spiral function
N = 500
t = np.linspace(0,20,N);
xtp = np.exp(t/10)*np.sin(5*t);
ytp = np.exp(t/10)*np.cos(5*t);

# matplotlib
fig1 = plt.figure(figsize=(6,5),dpi=100,facecolor='w',edgecolor='k')
ax1 = fig1.gca(projection='3d')
ax1.plot(xtp,ytp,t,linewidth=2)
ax1.set_xlabel("X",fontsize=12); ax1.set_ylabel("Y",fontsize=12); ax1.set_zlabel("t",fontsize=12)
plt.xlim(-8,8); plt.ylim(-8,8); ax1.set_zlim(0,20)
plt.savefig('exp10a.png',dpi=100,bbox_inches='tight')
plt.show()
```

```python
# using scatter
fig2 = plt.figure(figsize=(6,5),dpi=100,facecolor='w',edgecolor='k')
ax2 = fig2.gca(projection='3d')
ax2.scatter(xtp,ytp,t,c=plt.cm.jet(t/max(t)),marker="o",s=20)
ax2.set_xlabel("X",fontsize=12); ax2.set_ylabel("Y",fontsize=12); ax2.set_zlabel("t",fontsize=12)
plt.xlim(-8,8); plt.ylim(-8,8); ax2.set_zlim(0,20)
plt.savefig('exp10b.png',dpi=100,bbox_inches='tight')
plt.show()
```

```python
# using plot to plot section by section
fig3 = plt.figure(figsize=(6,5),dpi=100,facecolor='w',edgecolor='k')
ax3 = fig3.gca(projection='3d')
for i in range(len(t)-1):
    ax3.plot(xtp[i:i+2],ytp[i:i+2],t[i:i+2],color=plt.cm.jet(t[i+1]/max(t)),linewidth=2)
ax3.set_xlabel("X",fontsize=12); ax3.set_ylabel("Y",fontsize=12); ax3.set_zlabel("t",fontsize=12)
plt.xlim(-8,8); plt.ylim(-8,8); ax3.set_zlim(0,20)
plt.savefig('exp10c.png',dpi=100,bbox_inches='tight')
plt.show()
```
