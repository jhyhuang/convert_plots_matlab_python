```python
# example 3

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# data - Gamma function
shape_values = [1,2,3,5,9]
scale_values = [2,2,2,1,0.5]
rndata = np.zeros([5,1000])
rnpdf = np.zeros([5,200])
x = np.linspace(0,20,200)
i=0
for a, b in zip(shape_values,scale_values):
    rndata[i,:] = gamma.rvs(a,loc=0,scale=b,size=1000)
    rnpdf[i,:] = gamma.pdf(x,a,scale=b)
    i=i+1
    
# figure
fig = plt.figure(figsize=(5,3),dpi=100,facecolor='w',edgecolor='k')
cmap = ["k","g","r","b","y"]
for i in range(len(cmap)):
    plt.hist(rndata[i,:],bins=50,alpha=0.3,color=cmap[i],density=True)
    plt.plot(x,rnpdf[i],cmap[i],linewidth=2)
plt.xlabel('Observations',size=12)
plt.ylabel('Probability Density',size=12)
plt.title('Gamma distribution',size=16)
plt.savefig('exp03.png',dpi=100,bbox_inches='tight')
```
