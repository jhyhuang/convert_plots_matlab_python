```python
# example 2

import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat

# load data
data = loadmat('Data_exp02.mat')

# step 1: original
fig = plt.figure(figsize=(5,3),dpi=100,facecolor='w',edgecolor='k')
hFit = plt.plot(np.squeeze(data['xfit']),np.squeeze(data['yfit']))
hE = plt.errorbar(np.squeeze(data['xdata_m']),np.squeeze(data['ydata_m']),np.squeeze(data['ydata_s']))
hData = plt.plot(np.squeeze(data['xVdata']),np.squeeze(data['yVdata']))
hModel = plt.plot(np.squeeze(data['xmodel']),np.squeeze(data['ymodel']))
hCI1 = plt.plot(np.squeeze(data['xmodel']),np.squeeze(data['ymodelL']))
hCI2 = plt.plot(np.squeeze(data['xmodel']),np.squeeze(data['ymodelU']))
plt.title("Step 1: Original data",size=16)
plt.savefig('exp02a.png',dpi=100,bbox_inches='tight')
```

```python
# step 2: add markers
fig = plt.figure(figsize=(5,3),dpi=100,facecolor='w',edgecolor='k')
hFit = plt.plot(np.squeeze(data['xfit']),np.squeeze(data['yfit']))
hE = plt.errorbar(np.squeeze(data['xdata_m']),np.squeeze(data['ydata_m']),np.squeeze(data['ydata_s']))
hData = plt.plot(np.squeeze(data['xVdata']),np.squeeze(data['yVdata']))
hModel = plt.plot(np.squeeze(data['xmodel']),np.squeeze(data['ymodel']))
hCI1 = plt.plot(np.squeeze(data['xmodel']),np.squeeze(data['ymodelL']))
hCI2 = plt.plot(np.squeeze(data['xmodel']),np.squeeze(data['ymodelU']))
plt.setp(hFit,color=(0, 0, 0.5),linewidth=2)
plt.setp(hE,linestyle='None',linewidth=1,color=(0.3,0.3,0.3))
plt.setp(hData,linestyle='None',marker='.')
plt.setp(hModel,linestyle='--',linewidth=1.5,color='r');
plt.setp(hCI1,linestyle='-.',linewidth=1.5,color=(0.0,0.5,0.0))
plt.setp(hCI2,linestyle='-.',linewidth=1.5,color=(0.0,0.5,0.0))
plt.title('Step 2: Add marker properties',size=16)
plt.savefig('exp02b.png',dpi=100,bbox_inches='tight')
```

```python
# step 3: refine markers
fig = plt.figure(figsize=(5,3),dpi=100,facecolor='w',edgecolor='k')
hFit = plt.plot(np.squeeze(data['xfit']),np.squeeze(data['yfit']))
hE = plt.errorbar(np.squeeze(data['xdata_m']),np.squeeze(data['ydata_m']),np.squeeze(data['ydata_s']),
                  marker='o',markersize=6,markeredgecolor=(0.2,0.2,0.2),markerfacecolor=(0.7,0.7,0.7))
hData = plt.plot(np.squeeze(data['xVdata']),np.squeeze(data['yVdata']))
hModel = plt.plot(np.squeeze(data['xmodel']),np.squeeze(data['ymodel']))
hCI1 = plt.plot(np.squeeze(data['xmodel']),np.squeeze(data['ymodelL']))
hCI2 = plt.plot(np.squeeze(data['xmodel']),np.squeeze(data['ymodelU']))
plt.setp(hFit,color=(0, 0, 0.5),linewidth=2)
plt.setp(hE,linestyle='None',linewidth=1,color=(0.3,0.3,0.3))
plt.setp(hData,linestyle='None',marker='o',markersize=5,markeredgecolor='None',
         markerfacecolor=(.75,.75,1))
plt.setp(hModel,linestyle='--',linewidth=1.5,color='r')
plt.setp(hCI1,linestyle='-.',linewidth=1.5,color=(0.0,0.5,0.0))
plt.setp(hCI2,linestyle='-.',linewidth=1.5,color=(0.0,0.5,0.0))
plt.title('Step 3: Refine marker properties',size=16)
plt.savefig('exp02c.png',dpi=100,bbox_inches='tight')
```

```python
# step 4: Final
fig = plt.figure(figsize=(5,3),dpi=100,facecolor='w',edgecolor='k')
hFit = plt.plot(np.squeeze(data['xfit']),np.squeeze(data['yfit']))
hE = plt.errorbar(np.squeeze(data['xdata_m']),np.squeeze(data['ydata_m']),np.squeeze(data['ydata_s']),
                  marker='o',markersize=6,markeredgecolor=(0.2,0.2,0.2),markerfacecolor=(0.7,0.7,0.7),
                  label='Data ($\mu$ $\pm$ $\sigma)$')
hData = plt.plot(np.squeeze(data['xVdata']),np.squeeze(data['yVdata']))
hModel = plt.plot(np.squeeze(data['xmodel']),np.squeeze(data['ymodel']))
hCI1 = plt.plot(np.squeeze(data['xmodel']),np.squeeze(data['ymodelL']))
hCI2 = plt.plot(np.squeeze(data['xmodel']),np.squeeze(data['ymodelU']))
plt.setp(hFit,color=(0, 0, 0.5),linewidth=2,label='Fit ($C x^3$)')
plt.setp(hE,linestyle='None',linewidth=1,color=(0.3,0.3,0.3))
plt.setp(hData,linestyle='None',marker='o',markersize=5,markeredgecolor='None',
         markerfacecolor=(.75,.75,1))
plt.setp(hModel,linestyle='--',linewidth=1.5,color='r',label='Model ($C x^3$)')
plt.setp(hCI1,linestyle='-.',linewidth=1.5,color=(0.0,0.5,0.0),label='95% CI')
plt.setp(hCI2,linestyle='-.',linewidth=1.5,color=(0.0,0.5,0.0));
plt.xlabel('Length (m)'); plt.ylabel('Mass (kg)')
plt.text(10,800,r'C = 0.1 $\pm$ 0.008 (CI)')
plt.legend()
plt.xlim(8,26); plt.ylim(0,2500)
plt.title('Step 4: Final',size=16)
plt.savefig('exp02d.png',dpi=100,bbox_inches='tight')
```
