```python
# example 6

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# peaks funtion
def peaks(x,y):
    return 3*(1-x)**2*np.exp(-(x**2)-(y+1)**2) \
           -10*(x/5-x**3-y**5)*np.exp(-x**2-y**2) \
           -1/3*np.exp(-(x+1)**2-y**2)

# data
x = np.arange(-5.0,5.0,0.05)
y = np.arange(-5.0,5.0,0.05)
X, Y = np.meshgrid(x,y)
Z = peaks(X,Y)

# matplotlib
fig1 = plt.figure(figsize=(4,4),dpi=100,facecolor='w',edgecolor='k')
ax1 = fig1.add_subplot(111)
plt.contour(X,Y,Z,20)
ax1.set_aspect('equal')
plt.xlim(-5,5); plt.ylim(-5,5)
plt.title('Contour',fontsize=12)
plt.colorbar()
plt.savefig('exp06a.png',dpi=100,bbox_inches='tight')
plt.show()
```

```python
fig2 = plt.figure(figsize=(4,4),dpi=100,facecolor='w',edgecolor='k')
ax2 = fig2.add_subplot(111)
plt.contourf(X,Y,Z,20)
ax2.set_aspect('equal')
plt.xlim(-5,5); plt.ylim(-5,5)
plt.title('Contourf',fontsize=12)
plt.colorbar()
plt.savefig('exp06b.png',dpi=100,bbox_inches='tight')
plt.show()
```

```python
fig3 = plt.figure(figsize=(4,4),dpi=100,facecolor='w',edgecolor='k')
ax3 = fig3.add_subplot(111)
plt.pcolormesh(X,Y,Z)
ax3.set_aspect('equal')
plt.xlim(-5,5); plt.ylim(-5,5)
plt.title('Pcolor',fontsize=12)
plt.colorbar()
plt.savefig('exp06c.png',dpi=100,bbox_inches='tight')
plt.show()
```

```python
fig4 = plt.figure(figsize=(4,4),dpi=100,facecolor='w',edgecolor='k')
plt.imshow(Z,origin='lower',cmap='viridis',extent=[-5, 5, -5, 5])
plt.title('Image show',fontsize=12)
plt.colorbar()
plt.savefig('exp06d.png',dpi=100,bbox_inches='tight')
plt.show()
```

```python
fig5 = plt.figure(figsize=(4,4),dpi=100,facecolor='w',edgecolor='k')
plt.imshow(Z,origin='lower',cmap='viridis',extent=[-5, 5, -5, 5])
plt.title('Image show + contour',fontsize=12)
plt.colorbar()
plt.contour(X,Y,Z,colors='white')
plt.savefig('exp06e.png',dpi=100,bbox_inches='tight')
plt.show()
```
