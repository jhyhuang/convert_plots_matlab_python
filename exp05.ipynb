```python
# example 5

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# data
data = sns.load_dataset("mpg")

# matplotlib
fig = plt.figure(figsize=(6,3),dpi=100,facecolor='w',edgecolor='k')
plt.subplot(1,2,1)
plt.boxplot((data['mpg'],data['acceleration'])); plt.grid(True)
plt.xticks([1,2],['mpg','acceleration'])
plt.subplot(1,2,2)
plt.violinplot((data['mpg'],data['acceleration'])); plt.grid(True)
plt.xticks([1,2],['mpg','acceleration'])
plt.savefig('exp05a.png',dpi=100,bbox_inches='tight')
```

```python
# seaborn
fig = plt.figure(figsize=(6,3),dpi=100,facecolor='w',edgecolor='k')
plt.subplot(1,2,1)
sns.boxplot(data=(data['mpg'],data['acceleration']),linewidth=1,width=0.15)
plt.xticks([0,1],['mpg','acceleration'])
plt.grid(True)
plt.subplot(1,2,2)
sns.violinplot(data=(data['mpg'],data['acceleration']),linewidth=1,width=1.0)
plt.xticks([0,1],['mpg','acceleration'])
plt.grid(True)
plt.savefig('exp05b.png',dpi=100,bbox_inches='tight')
```

```python
# pandas
fig = plt.figure(figsize=(6,3),dpi=100,facecolor='w',edgecolor='k')
df = pd.DataFrame(data)
plt.subplot(1,2,1)
df.boxplot(column=['mpg','acceleration'])
plt.subplot(1,2,2)
plt.violinplot([df.mpg,df.acceleration])
plt.xticks([1,2],['mpg','acceleration'])
plt.grid(True)
plt.savefig('exp05c.png',dpi=100,bbox_inches='tight')
```
